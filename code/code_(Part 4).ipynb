# **Part 2: Data Preprocessing Techniques**

Handling missing values:

Perform mean imputation on all numerical features and mode imputation on all categorical features respectively

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

def mean_imputation(feature):

    mean_imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
    mean_imputer = mean_imputer.fit(feature)
    mean_imputed_data = mean_imputer.transform(feature)
    mean_imputed_data = pd.DataFrame(mean_imputed_data, columns = mean_imputer.get_feature_names_out())

    return mean_imputed_data

def median_imputation(feature):
    median_imputer = SimpleImputer(missing_values = np.nan, strategy = 'median')
    median_imputer = median_imputer.fit(feature)
    median_imputed_data = median_imputer.transform(feature)
    median_imputed_data = pd.DataFrame(median_imputed_data, columns = median_imputer.get_feature_names_out())

    return median_imputed_data

def mode_imputation(feature):

    mode_imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
    mode_imputer = mode_imputer.fit(feature)
    mode_imputed_data = mode_imputer.transform(feature)
    mode_imputed_data = pd.DataFrame(mode_imputed_data, columns = mode_imputer.get_feature_names_out())

    return mode_imputed_data

def constant_imputation(feature, constant_value):
    constant_imputer = SimpleImputer(missing_values = np.nan, strategy = 'constant', fill_value = constant_value)
    constant_imputer = constant_imputer.fit(feature)
    constant_imputed_data = constant_imputer.transform(feature)
    constant_imputed_data = pd.DataFrame(constant_imputed_data, columns = constant_imputer.get_feature_names_out())

    return constant_imputed_data
```

Normalization and standardization:

Apply StandardScaler on numerical features

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler

def standard_scaling(feature):

    standard_scaler = StandardScaler()
    numerical_features_standard_scaled = standard_scaler.fit_transform(feature)
    numerical_features_standard_scaled = pd.DataFrame(numerical_features_standard_scaled, columns = standard_scaler.get_feature_names_out())

    return numerical_features_standard_scaled

def minmax_scaling(feature):

    minmax_scaler = MinMaxScaler()
    numerical_features_minmax_scaled = minmax_scaler.fit_transform(feature)
    numerical_features_minmax_scaled = pd.DataFrame(numerical_features_minmax_scaled, columns = minmax_scaler.get_feature_names_out())

    return numerical_features_minmax_scaled

def robust_scaling(feature):

    robust_scaler = RobustScaler()
    numerical_features_robust_scaled = robust_scaler.fit_transform(feature)
    numerical_features_robust_scaled = pd.DataFrame(numerical_features_robust_scaled, columns = robust_scaler.get_feature_names_out())

    return numerical_features_robust_scaled
```

Encoding categorical variables:

Apply OneHotEncoder on nominal categorical features (i.e. C0, C1, C3, C4, C7, C8, C9, C10, C11, C12, C13, C15) and OrdinalEncoder on ordinal categorical features (i.e. C2, C5) respectively

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

def onehot_encoding(feature):

    onehot_encoder = OneHotEncoder(dtype = np.int64)
    categorical_features_onehot_encoded = onehot_encoder.fit_transform(feature).toarray()
    categorical_features_onehot_encoded = pd.DataFrame(categorical_features_onehot_encoded, columns = onehot_encoder.get_feature_names_out())

    return categorical_features_onehot_encoded

def ordinal_encoding(feature, category_order):

    ordinal_encoder = OrdinalEncoder(dtype = np.int64, categories = category_order)
    categorical_features_ordinal_encoded = ordinal_encoder.fit_transform(feature)
    categorical_features_ordinal_encoded = pd.DataFrame(categorical_features_ordinal_encoded, columns = ordinal_encoder.get_feature_names_out())

    return categorical_features_ordinal_encoded
```

*Feature selection using VarianceThreshold:*

```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import VarianceThreshold

def feature_selection_variance_threshold(feature):

    variance_threshold_selector = VarianceThreshold(threshold = 0.5)
    variance_threshold_selected_feature = variance_threshold_selector.fit_transform(feature)
    variance_threshold_selected_feature = pd.DataFrame(variance_threshold_selected_feature, columns = variance_threshold_selector.get_feature_names_out())

    return variance_threshold_selected_feature
```

```python
selected_features_after_variance_threshold = feature_selection_variance_threshold(combined_features_for_variance_threshold)

selected_features_after_variance_threshold
```

*Feature selection using SelectKBest:*

```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import mutual_info_regression

def feature_selecton_select_k_best_regression(feature, target):

    select_k_best_selector = SelectKBest(score_func = mutual_info_regression, k = 9)
    select_k_best_selected_feature = select_k_best_selector.fit_transform(feature, target)
    select_k_best_selected_feature = pd.DataFrame(select_k_best_selected_feature, columns = select_k_best_selector.get_feature_names_out())

    return select_k_best_selected_feature

def feature_selecton_select_k_best_classification(feature, target):

    select_k_best_selector = SelectKBest(score_func = mutual_info_classif, k = 9)
    select_k_best_selected_feature = select_k_best_selector.fit_transform(feature, target)
    select_k_best_selected_feature = pd.DataFrame(select_k_best_selected_feature, columns = select_k_best_selector.get_feature_names_out())

    return select_k_best_selected_feature
```

```python
import time
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

training_time_1 = np.zeros(3)
validation_accuracy_1 = np.zeros(3)
F1_score_1 = np.zeros(3)

y_predict_probability = np.empty([int(np.round(X1.shape[0]*0.2)), len(np.unique(y_classification_target))])

for a in range(3):

    print("\n\n")

    print("Trial ", a+1, ":")

    print("\n")

    X_train_1, X_validate_1, y_train_1, y_validate_1 = train_test_split(X1, y_classification_target, test_size = 0.2, random_state = 4211)
    print("Shape of X_train_1: ", X_train_1.shape)
    print("Shape of X_validate_1: ", X_validate_1.shape)
    print("Shape of y_train_1: ", y_train_1.shape)
    print("Shape of y_validate_1: ", y_validate_1.shape)

    print("\n")

    model_1 = SGDClassifier(loss = "log_loss", random_state = a+1, learning_rate = "constant", eta0 = 0.25)
    start = time.time()
    logistic_reg_1 = model_1.fit(X_train_1, y_train_1)
    stop = time.time()
    training_time_1[a] = stop - start
    print("Training time of trial ", a+1, ": ", training_time_1[a])

    print("\n")


    # store the y_predict_probability (i.e. probability estimates) for plotting ROC curve and calculating AUC value in Q18
    if (a == 2):
        y_predict_probability = logistic_reg_1.predict_proba(X_validate_1)


    prediction_1 = logistic_reg_1.predict(X_validate_1)
    validation_accuracy_1[a] = accuracy_score(y_validate_1, prediction_1)
    print("Validation accuracy of logistic_reg_1 of trial ", a+1, ": ", validation_accuracy_1[a])

    print("\n")

    F1_score_1[a] = f1_score(y_validate_1, prediction_1, pos_label = "success", average = "binary")
    print("F1 score of logistic_reg_1 of trial ", a+1, ": ", F1_score_1[a])

print("\n\n\n")

mean_training_time_1 = np.mean(training_time_1)
mean_validation_accuracy_1 = np.mean(validation_accuracy_1)
mean_F1_score_1 = np.mean(F1_score_1)
print("Mean training time of logistic_reg_1: ", mean_training_time_1)
print("Mean validation accuracy of logistic_reg_1: ", mean_validation_accuracy_1)
print("Mean F1 score of logistic_reg_1: ", mean_F1_score_1)

print("\n")

std_training_time_1 = np.std(training_time_1)
std_validation_accuracy_1 = np.std(validation_accuracy_1)
std_F1_score_1 = np.std(F1_score_1)
print("Standard deviation training time of logistic_reg_1: ", std_training_time_1)
print("Standard deviation validation accuracy of logistic_reg_1: ", std_validation_accuracy_1)
print("Standard deviation F1 score of logistic_reg_1: ", std_F1_score_1)
```
